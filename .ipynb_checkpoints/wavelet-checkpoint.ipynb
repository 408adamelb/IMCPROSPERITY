```python
import pandas as pd
import matplotlib.pyplot as plt
from wavelet import WaveletTransform, getExponent
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import r2_score
```

# Actual plot of starfruit

```python
df = pd.read_csv("prices_round_1_day_0.csv", delimiter=";")

df.loc[df['ask_volume_1'].isnull(), 'ask_volume_1'] = 0
df.loc[df['ask_volume_2'].isnull(), 'ask_volume_2'] = 0
df.loc[df['ask_volume_3'].isnull(), 'ask_volume_3'] = 0

df.loc[df['ask_price_1'].isnull(), 'ask_price_1'] = 0
df.loc[df['ask_price_2'].isnull(), 'ask_price_2'] = 0
df.loc[df['ask_price_3'].isnull(), 'ask_price_3'] = 0

df.loc[df['bid_volume_1'].isnull(), 'bid_volume_1'] = 0
df.loc[df['bid_volume_2'].isnull(), 'bid_volume_2'] = 0
df.loc[df['bid_volume_3'].isnull(), 'bid_volume_3'] = 0

df.loc[df['bid_price_1'].isnull(), 'bid_price_1'] = 0
df.loc[df['bid_price_2'].isnull(), 'bid_price_2'] = 0
df.loc[df['bid_price_3'].isnull(), 'bid_price_3'] = 0


starfruit = df[df['product'] == "STARFRUIT"]
```

# VWAP

```python
for i in range(6, 10000):
    tot_vol = starfruit[i]["bid_volume_1"] + starfruit[i]["bid_volume_2"] + starfruit[i]["bid_volume_3"]
    bid_prices = starfruit[i]["bid_volume_1"] * starfruit[i]["bid_price_1"] + starfruit[i]["bid_volume_2"] * starfruit[i]["bid_price_2"] + starfruit[i]["bid_volume_3"] * starfruit[i]["bid_price_3"]
    print(bid_prices/tot_vol)
    
```

# Wavelet transform moving average

```python
t = 6
window_size = 6
mses = []
full_signal = [np.array([0]) for _ in range(6)]
all_signals = []

for m in range(6, 10000):
    filtered = np.array(starfruit[m-t:m]["mid_price"])
    transform = WaveletTransform(waveletName="db4")
    coefficients = transform.dwt(filtered, level=6)
    reconstructed_signal = np.array(transform.idwt(coefficients, level=6))
    
    moving_avg_signal = np.convolve(reconstructed_signal, np.ones(window_size) / window_size, mode='valid')
    full_signal.append(moving_avg_signal)
    
full_signal_avg = np.stack(full_signal).flatten()
plt.plot(range(50,60), starfruit[50:60]["mid_price"])
plt.plot(range(50,60), full_signal_avg[50:60])
```

# Greedy wavelet transform (most recent)

```python
t = 6
min_price = [0 for _ in range(t)]
max_price = [0 for _ in range(t)]
pred_price = [0 for _ in range(t)]

min_t = 100
max_t = 110

for m in range(t,10000):
    filtered = np.array(starfruit[m-t:m]["mid_price"])
    transform = WaveletTransform(waveletName="db4")
    coefficients = transform.dwt(filtered, level=6)
    
    reconstructed_signal_min = np.array(transform.idwt(coefficients, level=6))
    if (m%6 == 0):
        cur_recon = np.zeros((6,))
        
    cur_recon[m%6:] += reconstructed_signal_min[:6-(m%6)]
    pred_price.append(cur_recon[m%6] / (m%6+1))
    
plt.plot(range(min_t,max_t), starfruit[min_t:max_t]["mid_price"])
plt.plot(range(min_t,max_t), pred_price[min_t:max_t])
```

```python
t = 6
mses = []
full_signal = [np.zeros(6,)]
for m in range(t, 10000, t):
    n=6
    filtered = np.array(starfruit[m-t:m]["ask_price_1"])
    timestep = np.arange(6)
    A = np.vstack((timestep,np.ones(len(timestep)))).T
    b, c = np.linalg.lstsq(A, filtered, rcond=None)[0]
    pred = timestep*b+c
    futures = starfruit[m+1:m+n+1]["ask_price_1"]
    
    if (len(pred) == len(futures)):
        mse = np.mean((pred - futures)**2)
        mses.append(mse)
    full_signal.append(pred)
    
full_signal = np.stack(full_signal).flatten()
print(r2_score(full_signal[10:20], starfruit[10:20]["ask_price_1"]))
plt.plot(range(40,52), starfruit[40:52]["ask_price_1"])
plt.plot(range(40,52), full_signal[40:52])
```
